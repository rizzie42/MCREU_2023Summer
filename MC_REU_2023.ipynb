Drop certain features

```python
df = df.drop(["patient_id", "death_date_source_id", 'source_id_x', 'diagnosis_date',
              'observation_date', 'derived_by_TriNetX', 'source_id_y',
              "tumor_site_code_system", "morphology_code_system", "morphology_code",
              "stage_code_system", "stage_code", "tumor_size",
              "number_of_lymph_nodes", "metastatic"], axis = 1)
```

```python
df
```

Preprocess Data (Discretization)

```python
# Sex: M = 0, F = 1
df['sex'] = df['sex'].replace(['M', 'F'], [0, 1])
```

```python
# Race: Unknown = 0, White = 1, Asian = 2, Black or African American = 3, American Indian or Alaska Native = 4, Native Hawaiian or Other Pacific Islander = 5
df['race'] = df['race'].replace(['Unknown', 'White', 'Asian', 'Black or African American', 'American Indian or Alaska Native', 'Native Hawaiian or Other Pacific Islander'], [0, 1, 2, 3, 4, 5])
```

```python
# Ethnicity: Unknown = 0, Not Hispanic or Latino = 1, Hispanic or Latino = 2
df['ethnicity'] = df['ethnicity'].replace(['Unknown', 'Not Hispanic or Latino', 'Hispanic or Latino'], [0, 1, 2])
```

```python
# Marital_Status: Unknown = 0, Married = 1, Single = 2
df['marital_status'] = df['marital_status'].replace(['Unknown', 'Married', 'Single'], [0, 1, 2])
```

```python
# Age: 2023 - year of birth
df = df.rename(columns = {'year_of_birth': 'age'})
df['age'] = 2023 - df['age']

# Remove patients without age
df = df[df['reason_yob_missing'] == 'Present']
df = df.drop(['reason_yob_missing'], axis = 1)
```

```python
# Status: Dead = 0, Alive = 1
df = df.rename(columns = {'month_year_death': 'status'})

for i in df.status.unique():
  if math.isnan(i):
    pass

  else:
    df.loc[df['status'] == i, 'status'] = 0

df['status'] = df['status'].fillna(1)
```

```python
# Region: Unknown = 0, Northeast = 1, South = 2, Midwest = 3, West = 4, Ex-US = 5
df['patient_regional_location'] = df['patient_regional_location'].replace(['Unknown', 'Northeast', 'South', 'Midwest', 'West', 'Ex-US'], [0, 1, 2, 3, 4, 5])
```

```python
# Tumor: Other tumor = 0, Brain tumor = 1
df = df.replace({'tumor_site_code': r'^C70'}, {'tumor_site_code': 1}, regex=True)
df = df.replace({'tumor_site_code': r'^C71'}, {'tumor_site_code': 1}, regex=True)
df = df.replace({'tumor_site_code': r'^C72'}, {'tumor_site_code': 1}, regex=True)
df.loc[df['tumor_site_code'] != 1, 'tumor_site_code'] = 0
```

```python
df
```

Restrict Patient

```python
# Retain only US patients
df = df[df['patient_regional_location'] != 0]
df = df[df['patient_regional_location'] != 5]
```

```python
df = df.dropna()
```

Statistical Analysis (Rizzie)

```python
# Mean Value
mean_value = df.mean()
print(mean_value)
```

```python
# Geometric Mean Value
#geometric_mean_values = df.apply(lambda x: math.pow(x.prod(), 1/len(x)), axis=0)
#print(geometric_mean_values)

# The tumor_site_code column won't work with the gmean function, possibly due to incompatible data type
from scipy.stats import gmean
geometric_mean_values = df.iloc[:, :-1].apply(gmean, axis=0)
print(geometric_mean_values)
```

```python
# Percentage (calculate the coverage of each unique value in an attribute (except age))
# Example: Male: 51%, Female: 49%
column_percentages = (df.sum()/ df.sum().sum() )*100
print(column_percentages)
```

```python
# Percentage (split the age into 10 groups and calculate the coverage of each group)
# Example: 0-10: 1%, 11-20: 3%, etc.
age_groups = pd.cut(df['age'], bins = 10)
coverage = age_groups.value_counts()/ len(df)*100
print(coverage)
```

```python
# ANOVA / Kruskal - Wallis (continuous - categorical)
F, anova_p_value = f_oneway(df['status'], df['tumor_site_code'])
print(anova_p_value)
print(F)
```

```python
from scipy.stats.contingency import expected_freq
# Chi Square (categorical - categorical)
# Age is a continuous feature, not a categorical feature

contigency_table = pd.crosstab(df['age'], df['tumor_site_code'])
chi2, p_value, dof, expected_freq = chi2_contingency(contigency_table)
print(p_value)
```
